```python
class Graph:
	def __init__(self, startState, goalState):
		self.startState = eightpuzzle(startState)
		self.goalState = eightpuzzle(goalState)
		self.max_depth = 0
		self.final_states = []

	def iterative_deep_search(self, state, depth = 0):
		if depth == self.max_depth:
			if state == self.goalState:
				return ['REACHED GOAL STATE']
		else:
			for positions in ('UP', 'DOWN', 'LEFT', 'RIGHT'):
				new_state = state.move(positions)
				if new_state:
					print(f' => {depth + 1}, {positions}')
					path = self.iterative_deep_search(new_state, depth+1)
					if path:
						self.final_states.append(new_state.state)
						return [positions] + path

	def solve(self):
		if self.startState == self.goalState:
			print('\nState is already goal state\n')
			return True
		for i in range(1, 10000): 
			self.max_depth = i
			path = self.iterative_deep_search(self.startState)
			if path:
				self.final_states = self.final_states[::-1]
				print(f'\n\nFinal Path to Goal State => { " ".join(path[0:-1]) }\n')
				path = path[:len(path)-1]
				print(f'Start State => {self.startState.state}')
				for states in range(len(self.final_states)):
					print(f'{path[states]} => {self.final_states[states]}')
				return True	
		

class eightpuzzle:
	def __init__(self, state):
		self.state = state
		self.row_pos, self.col_pos = self.find_zero()

	def __eq__(self, new_state):
		return self.state == new_state.state

	def find_zero(self):
		for r_num, row in enumerate(self.state):
			for c_num, col in enumerate(row):
				if col == '0': return r_num, c_num

	def valid_pos(self, r_pos, c_pos):
		return r_pos < len(self.state) and r_pos >= 0 and \
				c_pos < len(self.state[0]) and c_pos >= 0

	def swap_zero(self, r_pos, c_pos):
		new_state = [list(s) for s in self.state]
		temp = new_state[self.row_pos][self.col_pos]
		new_state[self.row_pos][self.col_pos] = new_state[r_pos][c_pos]
		new_state[r_pos][c_pos] = temp
		return new_state

	def move(self, direction):
		if direction == 'UP':
			if self.valid_pos(self.row_pos-1, self.col_pos):
				return eightpuzzle(self.swap_zero(self.row_pos-1, self.col_pos))
		elif direction == 'DOWN':
			if self.valid_pos(self.row_pos+1, self.col_pos):
				return eightpuzzle(self.swap_zero(self.row_pos+1, self.col_pos))		
		elif direction == 'LEFT':
			if self.valid_pos(self.row_pos, self.col_pos-1):
				return eightpuzzle(self.swap_zero(self.row_pos, self.col_pos-1))
		elif direction == 'RIGHT':
			if self.valid_pos(self.row_pos, self.col_pos+1):
				return eightpuzzle(self.swap_zero(self.row_pos, self.col_pos+1))


startState = [['0', '1', '3'], ['4', '2', '5'], ['7', '8', '6']]
goalState = [['1', '2', '3'], ['4', '5', '6'], ['7', '8', '0']]

g = Graph(startState, goalState)
g.solve()
```
